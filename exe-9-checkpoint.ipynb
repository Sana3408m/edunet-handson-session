```python
#Write a Python program to draw a scatter plot comparing two subject marks of Mathematics and Science. Use marks of 10 students
import matplotlib.pyplot as plt

# Marks of 10 students in Mathematics and Science
math_marks = [75, 80, 60, 85, 55, 70, 65, 90, 82, 78]
science_marks = [65, 75, 50, 80, 45, 60, 55, 85, 72, 68]

# Plotting a scatter plot for Mathematics vs. Science marks
plt.figure(figsize=(8, 6))
plt.scatter(math_marks, science_marks, color='blue', label='Marks')
plt.title('Scatter Plot of Mathematics vs. Science Marks')
plt.xlabel('Mathematics Marks')
plt.ylabel('Science Marks')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#:** Write a Python program to draw a scatter graph taking a random distribution in X and Y and plotted against each other
import matplotlib.pyplot as plt
import numpy as np

# Generating random data for x and y coordinates
np.random.seed(0)  # Setting seed for reproducibility
x = np.random.randn(100)  # Random x values from a standard normal distribution
y = np.random.randn(100)  # Random y values from a standard normal distribution

# Plotting a scatter plot
plt.figure(figsize=(8, 6))
plt.scatter(x, y, color='red', alpha=0.7)
plt.title('Scatter Plot of Random Distribution')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid(True)
plt.show()
```

```python
#Write a Python programming to create a pie chart with a title of the popularity of programming Languages
import matplotlib.pyplot as plt

# Programming languages and their popularity (replace this with your data)
languages = ['Python', 'Java', 'JavaScript', 'C#', 'C++']
popularity = [50, 30, 25, 20, 15]  # Popularity percentages or counts

# Creating a pie chart
plt.figure(figsize=(8, 6))
plt.pie(popularity, labels=languages, autopct='%1.1f%%', startangle=140)
plt.title('Popularity of Programming Languages')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle
plt.show()
```

```python
#Write a Python programming to display a bar chart of the popularity of programming Languages
import matplotlib.pyplot as plt

# Programming languages and their popularity (replace this with your data)
languages = ['Python', 'Java', 'JavaScript', 'C#', 'C++']
popularity = [50, 30, 25, 20, 15]  # Popularity percentages or counts

# Creating a bar chart
plt.figure(figsize=(8, 6))
plt.bar(languages, popularity, color='skyblue')
plt.xlabel('Programming Languages')
plt.ylabel('Popularity')
plt.title('Popularity of Programming Languages')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.grid(axis='y')  # Display gridlines on y-axis
plt.tight_layout()  # Adjust layout to prevent clipping of labels
plt.show()
```

```python
#Write a Python program to create multiple plots
import matplotlib.pyplot as plt
import numpy as np

# Generating data for plots
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)

# Creating multiple plots (subplots) in a single figure
fig, axs = plt.subplots(3, figsize=(8, 10))  # 3 subplots arranged in a vertical layout

# Plotting on each subplot
axs[0].plot(x, y1, color='blue')
axs[0].set_title('Sin(x)')
axs[0].set_xlabel('x')
axs[0].set_ylabel('sin(x)')

axs[1].plot(x, y2, color='red')
axs[1].set_title('Cos(x)')
axs[1].set_xlabel('x')
axs[1].set_ylabel('cos(x)')

axs[2].plot(x, y3, color='green')
axs[2].set_title('Tan(x)')
axs[2].set_xlabel('x')
axs[2].set_ylabel('tan(x)')

# Adjust layout to prevent overlap
plt.tight_layout()
plt.show()
```

```python
#Write a Python program to plot two or more lines with different styles
import matplotlib.pyplot as plt
import numpy as np

# Generating data for plots
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# Plotting multiple lines with different styles
plt.figure(figsize=(8, 6))

# Plotting y1 (sin(x)) with a solid blue line and markers
plt.plot(x, y1, linestyle='-', color='blue', label='Sin(x)')

# Plotting y2 (cos(x)) with a dashed red line and markers
plt.plot(x, y2, linestyle='--', color='red', label='Cos(x)')

# Adding labels, title, legend, and grid
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Multiple Lines with Different Styles')
plt.legend()
plt.grid(True)

plt.show()
```

```python
#Write a Python program to draw line charts of the financial data of Alphabet Inc. between October 3, 2016 to October 7, 2016.
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Generating dates from October 3, 2016 to October 7, 2016
dates = pd.date_range('2016-10-03', '2016-10-07')

# Generating random financial data for demonstration purposes
np.random.seed(0)
closing_prices = np.random.randint(400, 500, size=len(dates))

# Creating a DataFrame with dates and closing prices
financial_data = pd.DataFrame({'Date': dates, 'Closing Price': closing_prices})

# Plotting line chart for financial data of Alphabet Inc.
plt.figure(figsize=(8, 6))
plt.plot(financial_data['Date'], financial_data['Closing Price'], marker='o', linestyle='-', color='blue')
plt.title('Financial Data of Alphabet Inc. (Oct 3-7, 2016)')
plt.xlabel('Date')
plt.ylabel('Closing Price')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()

plt.show()
```
