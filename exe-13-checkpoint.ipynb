```python
#Display basic statistical metrics
import pandas as pd

# Sample DataFrame
data = {
    'A': [15, 18, 20, 25, 30],
    'B': [11, 22, 33, 44, 55],
    'C': [10, 20, 30, 40, 50]
}

# Creating DataFrame
df = pd.DataFrame(data)
```

```python
#Display basic statistical metrics
```

```python
# Mean of each column
mean_values = df.mean()
print("Mean:")
print(mean_values)
```

```python
# Median of each column
median_values = df.median()
print("\nMedian:")
print(median_values)
```

```python
# Mode of each column
mode_values = df.mode()
print("\nMode:")
print(mode_values)
```

```python
# Standard deviation of each column
std_values = df.std()
print("\nStandard Deviation:")
print(std_values)
```

```python
# Minimum and maximum of each column
min_values = df.min()
max_values = df.max()
print("\nMinimum:")
print(min_values)
print("\nMaximum:")
print(max_values)
```

```python
#Generate a correlation matrix to analyze the relationships between features
```

```python
# Generating correlation matrix
correlation_matrix = df.corr()

# Displaying correlation matrix
print("Correlation Matrix:")
print(correlation_matrix)
```

```python
#Create a histogram plot to visualize the distribution of the data
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Creating a sample DataFrame
data = {
    'Values': np.random.normal(loc=0, scale=1, size=1000)  # Generating random normal distribution data
}

# Creating DataFrame
df = pd.DataFrame(data)

# Plotting histogram
df.hist(column='Values', bins=30, grid=False, color='blue', figsize=(8, 6))
plt.title('Histogram of Values')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.show()
```

```python
#Create a bar, pie chart plot to visualization of the data
import pandas as pd
import matplotlib.pyplot as plt

# Sample data
data = {
    'Category': ['A', 'B', 'C', 'D'],
    'Values': [25, 40, 30, 55]
}

# Creating DataFrame
df = pd.DataFrame(data)

# Bar chart
plt.figure(figsize=(8, 5))
plt.bar(df['Category'], df['Values'], color='skyblue')
plt.xlabel('Category')
plt.ylabel('Values')
plt.title('Bar Chart')
plt.show()

# Pie chart
plt.figure(figsize=(6, 6))
plt.pie(df['Values'], labels=df['Category'], autopct='%1.1f%%', startangle=90, colors=['gold', 'yellowgreen', 'lightcoral', 'lightskyblue'])
plt.title('Pie Chart')
plt.axis('equal')
plt.show()
```
