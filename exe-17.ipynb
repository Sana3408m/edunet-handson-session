```python
#Implement linear regression for given data \n", "\n", "Salary_Data.csv
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# Load the dataset
data = pd.read_csv('Salary_Data.csv')

# Separate the independent and dependent variables
X = data['YearsExperience'].values.reshape(-1, 1)  # Independent variable
y = data['Salary']  # Dependent variable

# Create a linear regression model
model = LinearRegression()

# Fit the model to the data (training the model)
model.fit(X, y)

# Get the coefficients and intercept
coefficients = model.coef_
intercept = model.intercept_

print(f"Coefficient: {coefficients[0]}")
print(f"Intercept: {intercept}")

# Predict using the model
y_pred = model.predict(X)

# Plotting the original data and the regression line
plt.scatter(X, y, color='blue', label='Actual data')
plt.plot(X, y_pred, color='red', label='Regression line')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.title('Linear Regression: Salary Prediction based on Experience')
plt.legend()
plt.show()
```
